```python
import pandas as pd
df=pd.read_csv("Merge.csv")
df.head()
```

```python
df['Joint3_Torque'] = df['Joint3_Torque'].apply(lambda x: x if isinstance(x, tuple) and len(x) == 2 else (0, 0))

# Splitting the tuple into separate columns
df[['Joint3_Torque1', 'Joint3_Torque2']] = pd.DataFrame(df['Joint3_Torque'].tolist(), index=df.index)
df.drop('Joint3_Torque', axis=1, inplace=True)

# Verifying the split
print(df.head())
```

```python
df['Label'] = df['Label'].map({'rectangle': 1})

# Verifying the result
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Label'] = label_encoder.fit_transform(df['Label'])

# Verifying the result
print(df.head())
```

```python

X = df.drop(columns=['Joint1_Torque','Joint4_Torque','Label'], axis=1)
y = df['Label']
X.shape,y.shape
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import  RobustScaler
robust_scaler = RobustScaler()
X_robust = robust_scaler.fit_transform(X)
X_robust
```

```python
X_scaled_robust = pd.DataFrame(X_robust, columns=X.columns)
print("\nRobust Scaled DataFrame:")
X_scaled_robust.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_robust, y, test_size=0.2, random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Training the linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Predicting on the testing set
y_pred_linear = linear_reg.predict(X_test)


# Calculating R-squared score
r2_linear = r2_score(y_test, y_pred_linear)
print("Linear Regression R-squared Score:", r2_linear)
```

# **The R-squared score ranges from 0 to 1, where a score closer to 1 indicates that the model explains a larger proportion of the variance in the target variable and thus is better at making predictions.**

```python
from sklearn.tree import DecisionTreeRegressor

# Training the decision tree model
decision_tree = DecisionTreeRegressor(random_state=42)
decision_tree.fit(X_train, y_train)

# Predicting on the testing set
score_tree = decision_tree.score(X_test, y_test)
print("Decision Tree Score:", score_tree)
```

# **the model explains 100% of the variance in the target variable based on the features provided.**
